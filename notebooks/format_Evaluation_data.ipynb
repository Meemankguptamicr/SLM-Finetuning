```python
import pandas as pd
import ast
df = pd.read_csv(r"C:\Users\meemankgupta\Videos\SLM-Finetuning\notebooks\SLM_FINETUNING_BENCMARK_SLM.csv")

def convert_to_dict(val):
    if isinstance(val, str):
        try:
            return ast.literal_eval(val)  # Convert string to dictionary
        except (ValueError, SyntaxError):
            return None
    return val

# Apply the conversion function to all relevant columns
df['relevance'] = df['relevance'].apply(convert_to_dict)
df['groundedness'] = df['groundedness'].apply(convert_to_dict)
df['coherence'] = df['coherence'].apply(convert_to_dict)
df['fluency'] = df['fluency'].apply(convert_to_dict)
df['rouge'] = df['rouge'].apply(convert_to_dict)
df['gleu'] = df['gleu'].apply(convert_to_dict)
df['bleu'] = df['bleu'].apply(convert_to_dict)
df['meteor'] = df['meteor'].apply(convert_to_dict)
df['similarity'] = df['similarity'].apply(convert_to_dict)
df['f1'] = df['f1'].apply(convert_to_dict)

# Helper function to safely extract values from a dictionary
def safe_extract(d, key):
    if isinstance(d, dict) and key in d:
        return d[key]
    return None

# Extract numerical values from dictionaries into separate columns
df['relevance_score'] = df['relevance'].apply(lambda x: safe_extract(x, 'gpt_relevance'))
df['groundedness_score'] = df['groundedness'].apply(lambda x: safe_extract(x, 'gpt_groundedness'))
df['coherence_score'] = df['coherence'].apply(lambda x: safe_extract(x, 'gpt_coherence'))
df['fluency_score'] = df['fluency'].apply(lambda x: safe_extract(x, 'gpt_fluency'))
df['rouge_precision'] = df['rouge'].apply(lambda x: safe_extract(x, 'rouge_precision'))
df['rouge_recall'] = df['rouge'].apply(lambda x: safe_extract(x, 'rouge_recall'))
df['rouge_f1_score'] = df['rouge'].apply(lambda x: safe_extract(x, 'rouge_f1_score'))
df['gleu_score'] = df['gleu'].apply(lambda x: safe_extract(x, 'gleu_score'))
df['bleu_score'] = df['bleu'].apply(lambda x: safe_extract(x, 'bleu_score'))
df['meteor_score'] = df['meteor'].apply(lambda x: safe_extract(x, 'meteor_score'))
df['similarity_score'] = df['similarity'].apply(lambda x: safe_extract(x, 'gpt_similarity'))
df['f1_score'] = df['f1'].apply(lambda x: safe_extract(x, 'f1_score'))

# Drop the original dictionary columns (optional)
df = df.drop(columns=['relevance', 'groundedness', 'coherence', 'fluency', 'rouge', 'gleu', 'bleu', 'meteor', 'similarity', 'f1'])
```
